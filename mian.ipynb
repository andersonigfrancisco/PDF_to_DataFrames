```python
import pandas as pd
import os
import tabula

# Define os caminhos
base_path = os.path.join(os.getcwd(), 'docs', 'Source', 'B15', 'DDR')

# Lista todos os arquivos PDF no diretório especificado
pdf_files = [f for f in os.listdir(base_path) if f.lower().endswith('.pdf')]

# Páginas a serem extraídas (lista vazia para extrair todas as páginas)
pages_to_extract = []

specified_dfs = []

for pdf_file in pdf_files:
    pdf_path = os.path.join(base_path, pdf_file)
    try:
        dfs_list = []
        if pages_to_extract:
            for page_num in pages_to_extract:
                # Lê a página especificada do arquivo PDF
                dfs = tabula.read_pdf(pdf_path, pages=page_num,
                                      relative_area=True,
                                      relative_columns=True)
                if dfs:
                    dfs_list.extend(dfs)  # Adiciona os DataFrames da página especificada à lista
                    print(f"Successfully read {len(dfs)} tables from page {page_num} of {pdf_file}")
                else:
                    print(f"No tables found in page {page_num} of {pdf_file}")
        else:
            # Lê todas as páginas do arquivo PDF
            dfs = tabula.read_pdf(pdf_path, pages='all',
                                  relative_area=True,
                                  relative_columns=True)
            if dfs:
                dfs_list.extend(dfs)  # Adiciona todos os DataFrames do arquivo à lista
                print(f"Successfully read all tables from {pdf_file}")
            else:
                print(f"No tables found in {pdf_file}")
        
        specified_dfs.append((pdf_file, dfs_list))
    except Exception as e:
        print(f"Error reading {pdf_file}: {e}")

print(f"Total number of PDF files processed: {len(specified_dfs)}")
```

```python
# Display all DataFrames from each specified page of each PDF file
for pdf_file, dfs_list in specified_dfs:
    print(f"\nTables from {pdf_file}:\n")
    for idx, df in enumerate(dfs_list):
        print(f"Table {idx + 1} from {pdf_file}:\n")
        display(df)  # Use display() if in Jupyter Notebook, otherwise use print(df)
```
